# 2. Variables

Una variable es un espacio de memoria donde guardamos un dato para poder usarlo más adelante.

En Python no hace falta declarar el tipo por adelantado.  
Se escribe el nombre de la variable, el signo igual `=` y el valor.

### Recomendaciones para nombrar variables
- usar nombres claros
- no usar espacios
- usar guion bajo `_` si hace falta separar palabras
- evitar nombres poco descriptivos como `x` o `a` cuando el contexto no es obvio

Ejemplos de buenos nombres:
- `nombre`
- `edad`
- `precio_final`
- `cantidad_alumnos`

```python
nombre = "Ana"
edad = 25
altura = 1.68
es_estudiante = True

print(nombre)
print(edad)
print(altura)
print(es_estudiante)
```

# 5. Operadores

Los operadores permiten trabajar con datos.

## Operadores aritméticos
- `+` suma
- `-` resta
- `*` multiplicación
- `/` división
- `//` división entera
- `%` resto
- `**` potencia

## Operadores relacionales
Comparan valores y devuelven `True` o `False`.

- `==` igual
- `!=` distinto
- `>` mayor
- `<` menor
- `>=` mayor o igual
- `<=` menor o igual

## Operadores lógicos
- `and`
- `or`
- `not`

```python
a = 10
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / b)
print("División entera:", a // b)
print("Resto:", a % b)
print("Potencia:", a ** b)

print("a es mayor que b:", a > b)
print("a es igual a b:", a == b)
print("a es distinto de b:", a != b)

edad = 20
tiene_dni = True

print("Puede entrar:", edad >= 18 and tiene_dni)
```

# 6. Conversión de tipos

A veces necesitamos transformar un dato de un tipo a otro.

Funciones frecuentes:
- `int()`
- `float()`
- `str()`
- `bool()`

Esto es muy común cuando usamos `input()`, porque el dato ingresado llega como texto.

```python
numero_texto = "25"
numero_entero = int(numero_texto)

print(numero_texto, type(numero_texto))
print(numero_entero, type(numero_entero))

precio = 99.9
precio_texto = str(precio)

print(precio_texto, type(precio_texto))
```

# 7. Estructuras condicionales

Las condicionales permiten tomar decisiones.

## `if`
Ejecuta un bloque solo si la condición se cumple.

## `if - else`
Si la condición es verdadera, ejecuta un bloque.  
Si no, ejecuta otro.

## `if - elif - else`
Sirve para evaluar varias condiciones en orden.

```python
edad = 17

if edad >= 18:
    print("Es mayor de edad")
else:
    print("Es menor de edad")
```

```python
nota = 8

if nota >= 9:
    print("Excelente")
elif nota >= 7:
    print("Aprobado")
elif nota >= 4:
    print("Recupera")
else:
    print("Desaprobado")
```

# 8. Bucles o ciclos

Los bucles permiten repetir instrucciones.

## `for`
Se usa cuando sabemos cuántas veces queremos repetir algo o cuando recorremos una colección.

## `while`
Se usa cuando queremos repetir mientras se cumpla una condición.

```python
print("Ejemplo con for")
for i in range(5):
    print("Valor:", i)
```

```python
print("Ejemplo con while")
contador = 1

while contador <= 5:
    print("Contador:", contador)
    contador += 1
```

# 9. `range()`

La función `range()` genera una secuencia de números.

Formas comunes:
- `range(5)` → 0, 1, 2, 3, 4
- `range(2, 6)` → 2, 3, 4, 5
- `range(1, 10, 2)` → 1, 3, 5, 7, 9

Se usa muchísimo con `for`.

```python
for numero in range(2, 11, 2):
    print(numero)
```

# 10. Funciones

Una función es un bloque de código que realiza una tarea específica.  
Sirve para organizar, reutilizar y hacer más claro el programa.

## Ventajas de usar funciones
- evita repetir código
- mejora la legibilidad
- facilita pruebas y correcciones
- permite dividir un problema grande en partes pequeñas

Se definen con `def`.

```python
def saludar():
    print("Hola, bienvenidos")

saludar()
```

```python
def saludar_persona(nombre):
    print("Hola", nombre)

saludar_persona("Lucía")
saludar_persona("Pedro")
```

```python
def sumar(a, b):
    resultado = a + b
    return resultado

total = sumar(4, 7)
print("Resultado:", total)
```

# 11. Parámetros y retorno

## Parámetros
Son los datos que la función recibe para trabajar.

## `return`
Permite devolver un valor desde la función.

No todas las funciones devuelven algo.  
Algunas solo muestran información con `print()`.

```python
def calcular_promedio(n1, n2, n3):
    return (n1 + n2 + n3) / 3

promedio = calcular_promedio(8, 7, 9)
print("Promedio:", promedio)
```

# 12. Cadenas de texto (`str`)

Un string es una secuencia de caracteres.

Se puede:
- mostrar
- concatenar
- repetir
- recorrer
- acceder por posición
- usar métodos útiles

Recordá: en programación, muchas estructuras empiezan a contar desde 0.

```python
mensaje = "Python"

print(mensaje)
print("Primer carácter:", mensaje[0])
print("Último carácter:", mensaje[-1])
print("Cantidad de caracteres:", len(mensaje))
print("Concatenación:", mensaje + " es genial")
print("Repetición:", mensaje * 2)
```

```python
texto = "hola mundo"

print(texto.upper())
print(texto.capitalize())
print(texto.replace("mundo", "clase"))
print("hola" in texto)
```

# 13. Listas

Las listas son estructuras de datos que permiten guardar **varios elementos en un mismo lugar**.

Características:
- se escriben con corchetes `[]`
- pueden guardar distintos tipos de datos
- son ordenadas
- se pueden modificar

Las listas son una de las estructuras más usadas en Python.

```python
frutas = ["manzana", "banana", "pera"]

print(frutas)
print(frutas[0])
print(frutas[1])
print("Cantidad:", len(frutas))
```

```python
frutas = ["manzana", "banana", "pera"]

frutas.append("naranja")
print("Después de append:", frutas)

frutas.insert(1, "kiwi")
print("Después de insert:", frutas)

frutas.remove("banana")
print("Después de remove:", frutas)

eliminada = frutas.pop()
print("Elemento eliminado con pop:", eliminada)
print("Lista final:", frutas)
```

# 14. Recorrer listas

Una de las operaciones más comunes es recorrer una lista para leer sus elementos o procesarlos.

```python
numeros = [10, 20, 30, 40]

for numero in numeros:
    print("Valor:", numero)
```

```python
numeros = [10, 20, 30, 40]

for i in range(len(numeros)):
    print("Posición:", i, "- Valor:", numeros[i])
```

# 15. Operaciones comunes con listas

Podemos:
- sumar elementos numéricos
- buscar si un valor existe
- contar repeticiones
- ordenar
- invertir

Algunas funciones y métodos útiles:
- `sum()`
- `max()`
- `min()`
- `sorted()`
- `.sort()`
- `.count()`

```python
numeros = [8, 3, 10, 3, 6]

print("Lista original:", numeros)
print("Suma:", sum(numeros))
print("Mayor:", max(numeros))
print("Menor:", min(numeros))
print("Cantidad de veces que aparece 3:", numeros.count(3))
print("Ordenada:", sorted(numeros))
```

# 16. Tuplas

Las tuplas son parecidas a las listas, pero **no se pueden modificar** después de crearse.

Características:
- se escriben con paréntesis `()`
- son ordenadas
- son inmutables

Se usan cuando queremos que los datos no cambien.

```python
coordenadas = (10, 20)

print(coordenadas)
print("X:", coordenadas[0])
print("Y:", coordenadas[1])
```

# 17. Diccionarios

Los diccionarios guardan datos en pares:
**clave : valor**

Ejemplo:
- `"nombre"` → `"Ana"`
- `"edad"` → `25`

Características:
- se escriben con llaves `{}`
- permiten acceder a un valor a partir de su clave
- son muy útiles para representar objetos del mundo real

```python
alumno = {
    "nombre": "Sofía",
    "edad": 22,
    "curso": "Python"
}

print(alumno)
print(alumno["nombre"])
print(alumno["edad"])
```

```python
alumno = {
    "nombre": "Sofía",
    "edad": 22,
    "curso": "Python"
}

alumno["edad"] = 23
alumno["ciudad"] = "Buenos Aires"

print(alumno)

for clave, valor in alumno.items():
    print(clave, ":", valor)
```

# 18. Conjuntos (`set`)

Los conjuntos son colecciones:
- sin orden fijo
- sin elementos repetidos

Son útiles cuando queremos eliminar duplicados o trabajar con pertenencia.

```python
numeros = {1, 2, 2, 3, 4, 4, 5}

print(numeros)

numeros.add(6)
print(numeros)

print(3 in numeros)
```

# 19. Listas, tuplas, diccionarios y conjuntos: comparación rápida

## Lista
- ordenada
- mutable
- permite repetidos

## Tupla
- ordenada
- inmutable
- permite repetidos

## Diccionario
- pares clave-valor
- mutable
- acceso por clave

## Conjunto
- no ordenado
- mutable
- no permite repetidos

```python
lista = [1, 2, 2, 3]
tupla = (1, 2, 2, 3)
diccionario = {"a": 1, "b": 2}
conjunto = {1, 2, 2, 3}

print("Lista:", lista)
print("Tupla:", tupla)
print("Diccionario:", diccionario)
print("Conjunto:", conjunto)
```

# 20. Slicing

El slicing permite obtener una parte de una secuencia, como una lista o un string.

Sintaxis:
`estructura[inicio:fin:paso]`

Recordá:
- el inicio se incluye
- el fin no se incluye

```python
numeros = [10, 20, 30, 40, 50, 60]

print(numeros[1:4])
print(numeros[:3])
print(numeros[3:])
print(numeros[::2])
print(numeros[::-1])
```

# 21. Listas anidadas

Una lista puede contener otras listas.  
Esto se usa mucho para representar tablas, matrices o estructuras con filas y columnas.

```python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriz)
print("Elemento en fila 2, columna 3:", matriz[1][2])

for fila in matriz:
    print(fila)
```

# 23. Ejemplo integrador

Este ejemplo combina varios temas:
- variables
- entrada y salida
- listas
- condicionales
- funciones
- estructuras de datos

```python
def calcular_promedio_lista(lista_notas):
    return sum(lista_notas) / len(lista_notas)

alumno = {
    "nombre": "Valentina",
    "notas": [8, 9, 7, 10]
}

promedio = calcular_promedio_lista(alumno["notas"])

print("Alumno:", alumno["nombre"])
print("Notas:", alumno["notas"])
print("Promedio:", promedio)

if promedio >= 7:
    print("Aprobado")
else:
    print("No aprobado")
```
