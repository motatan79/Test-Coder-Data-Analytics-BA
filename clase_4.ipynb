## Lambda functions

```python
def suma(n1:int, n2:int) -> float:
    return n1 + n2

# Al lado de lambda va los parámetros de la función y después de los : va lo que hace la función.
sumar = lambda x1, x2: x1 + x2 
print('la suma de 2 y 3 es', sumar(2,3))

# Condicionales dentro de una lambda
# <valor_si_true> if <condicion> else <valor_si_false>
clasificar = lambda x: "Par" if x % 2 == 0 else "Impar"
l = [3,4,5,6]
l1= list(map(clasificar, l))
l2= list(map(lambda x: "Par" if x % 2 == 0 else "Impar", l))
print(l1)
print(l2)
```

## Calcular suma y promedio en una misma lambda

```python
resumen = lambda lista: (sum(lista), sum(lista)/len(lista))
lista = [10,20,30]
resumen(lista)
```

```python
# sorted se usa para los diccioarios
# sorted(iterable, key=, reverse=True)

estudiantes = [
    {'nombre': 'Alan', 
     'numero_favorito' : 23        
    },
    {'nombre': 'Kemi', 
     'numero_favorito' : 18        
    },
    {'nombre': 'Mike', 
     'numero_favorito' : 21        
    },
    {'nombre': 'Pedro', 
     'numero_favorito' : 17        
    }
]
# Si queremos ordenar este diccionario por alguna key, tenemos que usar sorted

estudiantes_sorted = sorted(estudiantes,key = lambda estudiante : estudiante['numero_favorito'])
print(estudiantes_sorted)
# para ordernar por nombre
estudiantes_sorted_nombres = sorted(estudiantes, key = lambda estudiantes : estudiantes['nombre'], reverse=True)
estudiantes_sorted_nombres
```

```python
numeros = [1,2,3,4,5]

cuadrados = list(map(lambda numero : numero * numero, numeros))
print(cuadrados)

# Vamos a sumar 20 a todos 
numeros_mas_20 = list(map(lambda numero : numero + 20, numeros))
numeros_mas_20
```

## Filter

```python
# Por ejemplo para saber los numeros pares de una lista
numeros = [1,2,3,4,5]
numeros_pares = list(filter(lambda numero: numero % 2 == 0, numeros))
numeros_pares
```

```python
# Hacer una lista de nombres
# Buscar Carla

nombres = ['Carla', 'Pedro', 'Juan', 'Carla', 'Lucrecia', 'Carla']

nombre_carla = list(filter(lambda nombre : 'Carla' in nombre, nombres))
nombre_carla
```

```python
## REDUCE
from functools import reduce

numeros = [1,2,3,4,5]

sumatoria = reduce(lambda a,b : a + b, numeros)
sumatoria
```
