# EJERCICIO 3 - Funciones Normales y Funciones Lambda

## a. Definir una función normal llamada multiplicar_por_dos que tome un número y lo multiplique por 2. Utilizar esta función para crear una nueva lista llamada dobles, que contenga los dobles de los números en la lista numeros.

```python
def multiplicar_por_dos(x:int) -> float:
    return x*2 
     
dobles = [multiplicar_por_dos(x) for x in numeros]
print(dobles)
```

## b. Definir una función lambda que haga lo mismo y utilizarla para crear una nueva lista llamada dobles_lambda, que contenga los dobles de los números en la lista numeros.

```python
# Otra forma 
m = lambda x : x * 2
dobles_lambdas = list(map(lambda x: x*2, numeros))
dobles_lambdas
```

# EJERCICIO 4 - Comprenhensions 

## a. Utilizar list comprehensions para crear una lista llamada cuadrados que contenga los cuadrados de los números en la lista numeros.

```python
cuadrados = [x*x for x in numeros]
print(cuadrados)
```

## b. Utilizar dict comprehensions para crear un diccionario llamado cubos que contenga los números de la lista numeros como claves y sus cubos como valores.

```python
cubos = {x:x*x*x for x in numeros}
print(cubos)
```

```python
cubo = lambda x : x*x*x

cubos = {x: map(cubo, numeros) for x in numeros}
cubos
```

## 5 - Comparación de resultados
## a. Mostrar las listas dobles, dobles_lambda y cuadrados. Comparar sus resultados y argumentar cuál es el mejor método.

```python
print(dobles)
print(dobles_lambdas)
print(cuadrados)
```

¿Cuál es más eficiente?
Para tu caso, el mejor método es:

dobles = [multiplicar_por_dos(x) for x in numeros]
cuadrados = [x*x for x in numeros]
Por qué
list comprehension es:

rápido en CPython
claro y legible
evita la llamada extra a map y a una función anónima lambda
En cambio, dobles_lambdas = list(map(lambda x: x*2, numeros)):

funciona bien
pero añade map() y lambda, que son una capa extra
suele ser un poco menos eficiente en tiempo de ejecución que la comprensión equivalente
Conclusión
Para operaciones simples como multiplicar por 2 o elevar al cuadrado, list comprehension es el método más eficiente y más fácil de entender.
map(lambda ...) puede ser útil, pero no aporta ventaja aquí y es menos directo.
Si quieres eficiencia y claridad, preferiría las comprensiones de listas.

## b. Imprimir el diccionario cubos y comparar con la creación de un diccionario usando un método convencional.

```python
cubos = {x:x*x*x for x in numeros}
print(cubos)

cubos_convencional = {}
for x in numeros:
    cubos_convencional[x] = x*x*x
print(cubos_convencional)
```

Más eficaz
Para este caso, el método más eficaz es:

Por qué
Es más compacto y legible.
Se ejecuta con menos sobrecarga de Python porque la comprensión está optimizada internamente.
Evita crear el diccionario vacío y luego hacer múltiples asignaciones con cubos_convencional[x] = ....
El método convencional
Funciona igual.
Pero es más lento en general porque hace más pasos Python a Python.
Conclusión
dict comprehension es la mejor opción cuando quieres crear un diccionario a partir de una lista de manera directa.
El bucle tradicional es útil si necesitas lógica adicional dentro del ciclo, pero no es la opción más eficiente aquí.
