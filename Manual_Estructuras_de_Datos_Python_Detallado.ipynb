# 1. ¿Qué es un objeto en Python?

Uno de los conceptos más importantes de Python es que **todo es un objeto**.

Esto significa que casi cualquier valor con el que trabajamos en Python tiene:

- un **tipo**
- un **valor**
- una **identidad** en memoria
- comportamientos asociados

Por ejemplo, todos estos son objetos:

- `5`
- `3.14`
- `"Hola"`
- `[1, 2, 3]`
- `{"nombre": "Ana"}`
- una función
- una tupla
- un conjunto

## ¿Qué implica esto?

Implica que cada valor:
- pertenece a un tipo
- puede guardarse en una variable
- puede pasarse a una función
- puede devolverse desde una función
- puede tener métodos u operaciones asociadas

```python
# Creamos distintos objetos y los guardamos en variables
numero = 10
texto = "Python"
lista = [1, 2, 3]
tupla = (4, 5, 6)
diccionario = {"nombre": "Ana"}
conjunto = {7, 8, 9}

# Mostramos cada valor
print(numero)
print(texto)
print(lista)
print(tupla)
print(diccionario)
print(conjunto)

# Mostramos el tipo de cada objeto
print(type(numero))
print(type(texto))
print(type(lista))
print(type(tupla))
print(type(diccionario))
print(type(conjunto))
```

# 2. Tipo, valor e identidad

Podemos pensar cada objeto con tres ideas:

## Tipo
Indica qué clase de objeto es.

## Valor
Es el contenido del objeto.

## Identidad
Es la “huella” del objeto en memoria.  
Se puede consultar con `id()`.

Este tema es importante para entender asignación, referencias y mutabilidad.

```python
# Creamos una lista
lista1 = [10, 20, 30]

# Asignamos la misma lista a otra variable
lista2 = lista1

# Mostramos las listas
print("lista1:", lista1)
print("lista2:", lista2)

# Mostramos la identidad de cada una
# Como ambas variables apuntan al mismo objeto, el id será igual
print("id lista1:", id(lista1))
print("id lista2:", id(lista2))
```

# 4. Mutabilidad e inmutabilidad

## Mutable
Un objeto mutable puede cambiar su contenido después de creado.

Ejemplos:
- listas
- diccionarios
- conjuntos

## Inmutable
Un objeto inmutable no puede cambiar su contenido una vez creado.

Ejemplos:
- enteros
- floats
- strings
- tuplas
- booleanos

Este tema afecta cómo funcionan las referencias, las asignaciones y las copias.

```python
# Ejemplo de objeto mutable
numeros = [1, 2, 3]
print("Lista original:", numeros)

# Modificamos el mismo objeto
numeros.append(4)
print("Lista modificada:", numeros)
```

# 5. String como objeto inmutable

Cuando “modificamos” un string, en realidad se crea un nuevo objeto y la variable pasa a referenciarlo.

```python
# Creamos un string
texto = "Hola"
print("Texto original:", texto)

# Esto no cambia el string original internamente:
# crea un nuevo string y lo reasigna a la variable
texto = texto + " mundo"

# Mostramos el resultado
print("Texto luego de concatenar:", texto)
```

# 6. Listas

Las listas son una de las estructuras de datos más usadas.

## Características
- se escriben con `[]`
- son **ordenadas**
- son **mutables**
- permiten repetidos
- pueden guardar tipos distintos
- permiten acceso por índice

```python
# Creamos una lista
frutas = ["manzana", "banana", "pera"]

# Mostramos la lista completa
print("Lista completa:", frutas)

# Accedemos por índice
print("Primer elemento:", frutas[0])
print("Segundo elemento:", frutas[1])
print("Último elemento:", frutas[-1])

# Mostramos la cantidad de elementos
print("Cantidad:", len(frutas))
```

# 7. Operaciones con listas

Con una lista podemos:
- agregar
- insertar
- eliminar
- recorrer
- ordenar
- copiar

```python
# Creamos una lista
numeros = [10, 20, 30]

# Agregamos un elemento al final
numeros.append(40)  # append agrega al final

# Insertamos un elemento en una posición específica
numeros.insert(1, 15)  # insert(posicion, valor)

# Mostramos la lista
print("Luego de append e insert:", numeros)

# Eliminamos un valor específico
numeros.remove(20)  # remove elimina por valor

# Mostramos el resultado
print("Luego de remove:", numeros)

# Eliminamos el último elemento y lo guardamos
ultimo = numeros.pop()  # pop sin parámetro elimina el último

# Mostramos qué se eliminó y la lista final
print("Eliminado con pop:", ultimo)
print("Lista final:", numeros)
```

# 8. Recorrido de listas

Recorrer una lista significa procesar sus elementos uno por uno.

La forma más común es con `for`.

```python
# Lista de nombres
nombres = ["Ana", "Luis", "Carla", "Pedro"]

# Recorremos por elemento
for nombre in nombres:
    # nombre toma cada valor de la lista
    print("Nombre:", nombre)
```

# 9. Recorrido por índice

También podemos recorrer la lista usando posiciones.

```python
# Lista de notas
notas = [8, 6, 9, 10]

# Recorremos usando índices
for i in range(len(notas)):
    # i es la posición actual
    print("Posición:", i, "- Nota:", notas[i])
```

# 10. Referencias y listas mutables

Si dos variables apuntan a la misma lista, modificarla desde una variable afecta a la otra.

```python
# Creamos una lista
lista_a = [1, 2, 3]

# lista_b apunta al mismo objeto
lista_b = lista_a

# Modificamos usando lista_b
lista_b.append(4)

# Ambas variables reflejan el cambio
print("lista_a:", lista_a)
print("lista_b:", lista_b)
```

# 11. Copia de listas

Si queremos una copia independiente, podemos usar:

- `copy()`
- `list()`
- slicing `[:]`

```python
# Lista original
original = [10, 20, 30]

# Copia independiente
copia = original.copy()

# Modificamos la copia
copia.append(40)

# La original no cambia
print("Original:", original)
print("Copia:", copia)

# Verificamos que son objetos diferentes
print("id original:", id(original))
print("id copia:", id(copia))
```

# 12. Tuplas

Las tuplas son parecidas a las listas, pero tienen una diferencia fundamental:

## Son inmutables

## Características
- se escriben con `()`
- son ordenadas
- son inmutables
- permiten repetidos
- permiten acceso por índice

```python
# Creamos una tupla
coordenadas = (100, 200)

# Mostramos sus datos
print("Tupla:", coordenadas)
print("Primer valor:", coordenadas[0])
print("Segundo valor:", coordenadas[1])
print("Cantidad:", len(coordenadas))
```

# 13. Inmutabilidad de las tuplas

Podemos leer sus elementos, pero no modificarlos.

```python
# Creamos una tupla
datos = (1, 2, 3)

# Leemos un elemento
print(datos[0])

# La siguiente línea daría error si se descomenta
# datos[0] = 99

print("Las tuplas no permiten modificar sus elementos.")
```

# 14. Comparación entre listas y tuplas

## Lista
- mutable
- útil cuando los datos pueden cambiar

## Tupla
- inmutable
- útil cuando los datos no deberían cambiar

```python
lista = [1, 2, 3]
tupla = (1, 2, 3)

# Modificamos la lista
lista[0] = 100

# Mostramos resultados
print("Lista modificada:", lista)
print("Tupla sin cambios:", tupla)
```

# 15. Conjuntos (`set`)

Los conjuntos son útiles cuando queremos trabajar con elementos únicos.

## Características
- se escriben con `{}` si tienen elementos
- son mutables
- no permiten repetidos
- no garantizan orden fijo
- no usan índices

```python
# Creamos un conjunto con repetidos
valores = {1, 2, 2, 3, 4, 4, 5}

# Los repetidos desaparecen automáticamente
print("Conjunto:", valores)

# Agregamos un elemento
valores.add(6)

# Mostramos el conjunto
print("Luego de add:", valores)

# Verificamos pertenencia
print("¿Está el 3?:", 3 in valores)
print("¿Está el 10?:", 10 in valores)
```

# 16. Operaciones con conjuntos

Los conjuntos permiten operaciones muy útiles:
- unión
- intersección
- diferencia
- diferencia simétrica

```python
# Creamos dos conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Mostramos operaciones
print("Unión:", a | b)
print("Intersección:", a & b)
print("Diferencia a - b:", a - b)
print("Diferencia simétrica:", a ^ b)
```

# 17. Diccionarios

Los diccionarios almacenan datos como pares **clave: valor**.

## Características
- se escriben con llaves
- son mutables
- se accede por clave
- cada clave debe ser única
- son ideales para representar entidades con atributos

```python
# Creamos un diccionario
alumno = {
    "nombre": "Sofía",
    "edad": 22,
    "curso": "Python"
}

# Mostramos el diccionario completo
print("Diccionario completo:", alumno)

# Accedemos por clave
print("Nombre:", alumno["nombre"])
print("Edad:", alumno["edad"])
print("Curso:", alumno["curso"])
```

# 18. Modificar diccionarios

En un diccionario podemos:
- cambiar valores
- agregar claves nuevas
- eliminar claves

```python
# Creamos un diccionario
producto = {
    "nombre": "Teclado",
    "precio": 15000,
    "stock": 8
}

# Modificamos una clave existente
producto["precio"] = 18000

# Agregamos una nueva clave
producto["marca"] = "Logitech"

# Mostramos el resultado
print(producto)

# Eliminamos una clave
del producto["stock"]

# Mostramos el resultado final
print(producto)
```

# 19. Recorrido de diccionarios

Podemos recorrer un diccionario:
- por claves
- por valores
- por clave y valor

```python
# Creamos un diccionario
persona = {
    "nombre": "Lucía",
    "edad": 30,
    "ciudad": "Lanús"
}

# Recorremos claves
for clave in persona:
    print("Clave:", clave)

print("-----")

# Recorremos valores
for valor in persona.values():
    print("Valor:", valor)

print("-----")

# Recorremos clave y valor
for clave, valor in persona.items():
    print(clave, ":", valor)
```

# 21. Funciones normales

Una función normal se define con `def`.

Conviene usarla cuando:
- la lógica ocupa varias líneas
- queremos reutilizar código
- queremos darle un nombre claro a una tarea

```python
# Definimos una función normal
def cuadrado(numero):
    # Calculamos el cuadrado
    resultado = numero ** 2

    # Devolvemos el resultado
    return resultado

# Llamamos a la función
valor = cuadrado(5)

# Mostramos el resultado
print("Cuadrado:", valor)
```

# 22. Funciones lambda

Una función lambda es una función pequeña y anónima.

## Sintaxis
`lambda parámetros: expresión`

## Diferencias con una función normal

### Función normal
- usa `def`
- puede tener varias líneas
- suele ser más clara para lógica compleja

### Lambda
- es breve
- suele escribirse en una sola línea
- es útil para tareas pequeñas y puntuales

```python
# Creamos una lambda que calcula el cuadrado
cuadrado_lambda = lambda numero: numero ** 2

# La usamos como una función
print(cuadrado_lambda(6))

# Misma lógica con función normal
def cuadrado_normal(numero):
    return numero ** 2

print(cuadrado_normal(6))
```

# 23. Lambda con varios parámetros

Las lambdas también pueden recibir más de un parámetro.

```python
# Lambda que suma dos números
sumar = lambda a, b: a + b

# Mostramos el resultado
print("Suma:", sumar(10, 20))
```

# 24. ¿Cuándo conviene usar lambda?

Conviene usar lambda cuando:
- la operación es muy corta
- se usa como argumento de otra función
- queremos evitar definir una función completa para algo simple

No conviene cuando la lógica es larga o poco legible.

```python
# Lista de palabras
palabras = ["sol", "programacion", "luz", "python"]

# Ordenamos por longitud usando lambda
ordenadas = sorted(palabras, key=lambda palabra: len(palabra))

# Mostramos el resultado
print("Ordenadas por longitud:", ordenadas)
```

# 25. Comprensión de listas

La comprensión de listas permite crear listas de manera compacta.

## Sintaxis
`[expresión for elemento in secuencia]`

También puede incluir condición:
`[expresión for elemento in secuencia if condición]`

```python
# Lista de números
numeros = [1, 2, 3, 4, 5]

# Creamos una nueva lista con los cuadrados
cuadrados = [numero ** 2 for numero in numeros]

# Mostramos resultados
print("Original:", numeros)
print("Cuadrados:", cuadrados)
```

# 26. Comprensión de listas con filtro

Podemos usar una condición para quedarnos solo con algunos elementos.

```python
# Lista de números
numeros = [1, 2, 3, 4, 5, 6, 7, 8]

# Creamos una lista solo con números pares
pares = [numero for numero in numeros if numero % 2 == 0]

# Mostramos resultados
print("Original:", numeros)
print("Pares:", pares)
```

# 27. For tradicional vs comprensión de listas

La comprensión de listas puede reemplazar algunos bucles simples.

```python
# Forma tradicional
numeros = [1, 2, 3, 4]
dobles = []

for numero in numeros:
    # Agregamos el doble de cada número
    dobles.append(numero * 2)

print("Con for:", dobles)

# Forma con comprensión
dobles_compresion = [numero * 2 for numero in numeros]

print("Con comprensión:", dobles_compresion)
```

# 28. Comprensión de diccionarios

También podemos construir diccionarios de manera compacta.

## Sintaxis
`{clave: valor for elemento in secuencia}`

```python
# Lista de números
numeros = [1, 2, 3, 4, 5]

# Creamos un diccionario con número: cuadrado
diccionario_cuadrados = {numero: numero ** 2 for numero in numeros}

# Mostramos el resultado
print(diccionario_cuadrados)
```

# 29. Otro ejemplo de comprensión de diccionarios

Podemos usarla para transformar palabras en un diccionario de longitudes.

```python
# Lista de palabras
palabras = ["sol", "luna", "estrella", "mar"]

# Creamos un diccionario palabra: cantidad de letras
longitudes = {palabra: len(palabra) for palabra in palabras}

# Mostramos el resultado
print(longitudes)
```

# 30. Listas de diccionarios

Una lista de diccionarios representa una colección de elementos donde cada elemento tiene varios atributos.

Ejemplos:
- alumnos
- productos
- empleados
- películas

```python
# Lista de diccionarios con alumnos
alumnos = [
    {"nombre": "Ana", "nota": 8, "curso": "Python"},
    {"nombre": "Luis", "nota": 6, "curso": "Python"},
    {"nombre": "Carla", "nota": 9, "curso": "Python"}
]

# Mostramos la lista completa
print(alumnos)

print("-----")

# Recorremos la lista
for alumno in alumnos:
    # alumno es un diccionario en cada vuelta
    print("Nombre:", alumno["nombre"])
    print("Nota:", alumno["nota"])
    print("Curso:", alumno["curso"])
    print("-----")
```

# 31. Filtrar listas de diccionarios

Podemos mostrar solo los elementos que cumplan una condición.

```python
# Lista de productos
productos = [
    {"nombre": "Mouse", "precio": 12000, "stock": 10},
    {"nombre": "Teclado", "precio": 25000, "stock": 0},
    {"nombre": "Monitor", "precio": 180000, "stock": 4}
]

# Recorremos y filtramos por stock
for producto in productos:
    if producto["stock"] > 0:
        print("Producto disponible:", producto["nombre"])
```

# 32. Filtrar listas de diccionarios con comprensión

También podemos filtrar creando una nueva lista.

```python
# Lista de productos
productos = [
    {"nombre": "Mouse", "precio": 12000, "stock": 10},
    {"nombre": "Teclado", "precio": 25000, "stock": 0},
    {"nombre": "Monitor", "precio": 180000, "stock": 4}
]

# Nueva lista solo con productos disponibles
disponibles = [producto for producto in productos if producto["stock"] > 0]

# Mostramos el resultado
print(disponibles)
```

# 33. Transformar listas de diccionarios

A partir de una lista de diccionarios podemos construir otras estructuras.

```python
# Lista de alumnos
alumnos = [
    {"nombre": "Ana", "nota": 8},
    {"nombre": "Luis", "nota": 6},
    {"nombre": "Carla", "nota": 9}
]

# Lista con solo los nombres
nombres = [alumno["nombre"] for alumno in alumnos]

# Lista con notas aprobadas
notas_aprobadas = [alumno["nota"] for alumno in alumnos if alumno["nota"] >= 7]

# Mostramos resultados
print("Nombres:", nombres)
print("Notas aprobadas:", notas_aprobadas)
```

# 34. Método `get()` en diccionarios

Cuando usamos `diccionario["clave"]`, si la clave no existe, hay error.

Con `get()` podemos evitarlo y hasta indicar un valor por defecto.

```python
# Creamos un diccionario
persona = {"nombre": "María", "edad": 28}

# Accedemos a datos existentes
print(persona["nombre"])
print(persona.get("edad"))

# Clave inexistente
print(persona.get("ciudad"))

# Clave inexistente con valor por defecto
print(persona.get("ciudad", "No informada"))
```

# 36. Ejemplo integrador

En este ejemplo combinamos:
- lista de diccionarios
- función normal
- lambda
- comprensión de listas
- comprensión de diccionarios

```python
# Lista de alumnos
alumnos = [
    {"nombre": "Ana", "nota": 8},
    {"nombre": "Luis", "nota": 6},
    {"nombre": "Carla", "nota": 9},
    {"nombre": "Pedro", "nota": 7}
]

# Función normal para saber si aprueba
def esta_aprobado(nota):
    # Devuelve True si la nota es 7 o más
    return nota >= 7

# Filtramos aprobados
aprobados = [alumno for alumno in alumnos if esta_aprobado(alumno["nota"])]

# Lista solo con nombres aprobados
nombres_aprobados = [alumno["nombre"] for alumno in aprobados]

# Ordenamos por nota con lambda
ordenados_por_nota = sorted(alumnos, key=lambda alumno: alumno["nota"], reverse=True)

# Diccionario nombre: nota
notas_por_nombre = {alumno["nombre"]: alumno["nota"] for alumno in alumnos}

# Mostramos resultados
print("Aprobados:", aprobados)
print("Nombres aprobados:", nombres_aprobados)
print("Ordenados por nota:", ordenados_por_nota)
print("Notas por nombre:", notas_por_nombre)
```
